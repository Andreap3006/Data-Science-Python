#Importando la Base de Datos

```python
import pandas as pd
```

```python
datos = pd.read_csv('credit_card.csv')
datos.head()
```

```python
datos.info()
```

#Traduciendo variables

```python
#Conociendo nombres de columnas
datos.columns
```

```python
#Creando diccionario para renombrar columnas
dict_columnas ={
    'LIMIT_BAL': 'Limite',
    'CHECKING_ACCOUNT': 'Cuenta_corriente',
    'EDUCATION': 'Escolaridad',
    'MARRIAGE': 'Estado_civil',
    'AGE': 'Edad',
    'BILL_AMT': 'Valor_factura',
    'PAY_AMT': 'Valor_pagado',
    'DEFAULT': 'Moroso'
}
```

```python
#Creando nueva BBDD con columnas renombradas
tarjetas = datos.rename(columns = dict_columnas)
tarjetas.head()
```

```python
#Conociendo datos de la columna "Cuenta_corriente"
tarjetas.Cuenta_corriente.unique()
```

```python
#Creando diccionario para renombrar datos
dict_cuenta = {
    'Yes': 'Si',
    'No': 'No'
}
```

```python
#Renombrando valores de columna "Cuenta_corriente"
tarjetas.Cuenta_corriente = tarjetas.Cuenta_corriente.map(dict_cuenta)
tarjetas.head()
```

```python
#Conociendo datos de la columna "Escolaridad"
tarjetas.Escolaridad.unique()
```

```python
#Creando diccionario para renombrar datos
dict_escolaridad = {
    '2.University': '2.Universidad',
    '3.Graduate School': '3.Pos-graduaci√≥n',
    '1.High School': '1.Colegio'
}
```

```python
#Renombrando valores de columna "Escolaridad"
tarjetas.Escolaridad = tarjetas.Escolaridad.map(dict_escolaridad)
tarjetas.head()
```

```python
#Conociendo datos de la columna "Estado_civil"
tarjetas.Estado_civil.unique()
```

```python
#Creando diccionario para renombrar datos
dict_estado_civil = {
    'Married': 'Casado/a',
    'Single': 'Soltero/a'
}
```

```python
#Renombrando valores de columna "Estado_civil"
tarjetas.Estado_civil = tarjetas.Estado_civil.map(dict_estado_civil)
tarjetas.head()
```

**Regresi√≥n**

La l√≠nea de dispersi√≥n la podemos ver usando la funci√≥n lmplot (linear model plot), este gr√°fico nos mostrar√° tambi√©n un gr√°fico de dispersi√≥n y la l√≠nea de regresi√≥n que mejor se ajuste a la relaci√≥n entre las dos variables.

**Test de Hip√≥tesis**

Buscando entender c√≥mo se relaciona la variable del valor de factura con la variable de moroso y no moroso. Para conocer si estos dos grupos son estad√≠sticamente diferentes o no.

```python
sns.lmplot(x='iu', y='Valor_factura', data=tarjetas, col='Moroso');
```

```python
#Importando biblioteca scipy para calcular el p-value
from scipy.stats import ranksums
```

H<sup>null<sup>

La distribuci√≥n de los grupos moroso y no moroso es la misma

H<sup>alt<sup>

La distribuci√≥n de los grupos moroso y no moroso no es la misma

```python
#Separamos los grupos moroso y no moroso usando el valor de la factura como base
moroso = tarjetas.query("Moroso == 1").Valor_factura
moroso
```

```python
no_moroso = tarjetas.query("Moroso == 0").Valor_factura
no_moroso
```

```python
#Usamos la funci√≥n ranksums de scipy para calcular el p-value
resultado = ranksums(moroso, no_moroso)
```

```python
print("El p_value de nuestro test es {}".format(resultado.pvalue))
```

La distribuci√≥n del valor de factura entre los grupos moroso y no moroso no es igual, por lo que podemos aceptar la hip√≥tesis alternativa.
