**Test de Hip贸tesis**

Buscando entender c贸mo se relaciona la variable del valor de factura con la variable de moroso y no moroso. Para conocer si estos dos grupos son estad铆sticamente diferentes o no.

```python
sns.lmplot(x='iu', y='Valor_factura', data=tarjetas, col='Moroso');
```

```python
#Importando biblioteca scipy para calcular el p-value
from scipy.stats import ranksums
```

H<sup>null<sup>

La distribuci贸n de los grupos moroso y no moroso es la misma

H<sup>alt<sup>

La distribuci贸n de los grupos moroso y no moroso no es la misma

```python
#Separamos los grupos moroso y no moroso usando el valor de la factura como base
moroso = tarjetas.query("Moroso == 1").Valor_factura
moroso
```

```python
no_moroso = tarjetas.query("Moroso == 0").Valor_factura
no_moroso
```

```python
#Usamos la funci贸n ranksums de scipy para calcular el p-value
resultado = ranksums(moroso, no_moroso)
```

```python
print("El p_value de nuestro test es {}".format(resultado.pvalue))
```

La distribuci贸n del valor de factura entre los grupos moroso y no moroso no es igual, por lo que podemos aceptar la hip贸tesis alternativa.
